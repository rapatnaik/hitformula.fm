```python
import pandas as pd


# === Load files using absolute paths and .csv extension ===
songs = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_metadata/songs.csv', sep='\t')
artists = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_metadata/artists.csv', sep='\t')
albums = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_metadata/albums.csv', sep='\t')
releases = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_metadata/releases.csv', sep='\t')
tracks = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_metadata/tracks.csv', sep='\t')


song_pop = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_popularity/song_pop.csv', sep='\t')
song_chart = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_popularity/song_chart.csv', sep='\t')
album_pop = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_popularity/album_pop.csv', sep='\t')
album_chart = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_popularity/album_chart.csv', sep='\t')
artist_pop = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_popularity/artist_pop.csv', sep='\t')
artist_chart = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_popularity/artist_chart.csv', sep='\t')


lyrics = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_songfeatures/lyrics.csv', sep='\t')
acoustic = pd.read_csv('/Users/andrewchen/dsc106/music stuff/musicoset_songfeatures/acoustic_features.csv', sep='\t')
songs
```

```python
songs_features = (
    songs
      .merge(acoustic, on="song_id", how="left")   # add audio features
      .merge(lyrics,   on="song_id", how="left")   # add lyrics
)


songs_popularity = (
    songs_features
      .merge(song_pop,   on="song_id", how="left")   # year-end pop score
      .merge(song_chart, on="song_id", how="left")   # weekly chart stats
)

albums_popularity = (
    albums
      .merge(album_pop,   on="album_id", how="left")   # year-end pop score
      .merge(album_chart, on="album_id", how="left")   # weekly chart stats
)


tracks_enriched = (
    tracks                                              # gives album_id, track_number
      .merge(albums_popularity, on="album_id", how="left")
      .merge(releases,         on=["album_id"], how="left")  # release dates
)

artists_enriched = (
    artists
      .merge(artist_pop,   on="artist_id", how="left")   # year-end pop score
      .merge(artist_chart, on="artist_id", how="left")   # weekly chart stats
)
```
